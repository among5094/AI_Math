# 평균,분산,표준편차 사용해보기

### 대전의 일주일 날씨 데이터 사용하기

- 대전 2023년 07월 05일 수요일 ~ 7월 12일 수요일 날씨 데이터



```python
import numpy as np
import matplotlib.pyplot as plt

#대전의 시간별 기온 데이터 입력
list_tem = [29, 32, 26, 31, 32, 30, 29, 30]
Temperatures = np.array(list_tem)
print("대전 일주일 날씨:",Temperatures) #일주일치 날씨 데이터 입력 확인하기


#대전의 일주일 날씨 평균 구하기 -> np.mean() 사용
Temperatures_mean = np.mean(Temperatures, axis=0)
print("평균:", Temperatures_mean )


#분산(데이터가 얼마나 퍼져있는지) 구하기 -> np.var()
Temperatures_var = np.var(Temperatures, axis=0) 
print("분산:", Temperatures_var )


#분산은 제곱을 하기 때문에 값이 너무 커져서 감이 오지 않는다.
#표준편차(데이터와 비슷한 크기로 분산을 변화) 구하기 -> np.std()
Temperatures_std = np.std(Temperatures)
print("표준편차:", Temperatures_std )
plt.hist(Temperatures_std)
plt.show()

```
