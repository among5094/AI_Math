# 노이즈(에러)가 포함된 랜덤 직선 데이터 생성

- 일부러 에러를 포함시킴. 에러가 포함되었을 때의 직선을 보기위해

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(1, 10, 100) # <- 균일한 간격으로 값들을 생성하여 배열로 변환
error = np.random.normal(0, 1, 100) #0부터 1까지 100개의 데이터

y =2*x + 5 + error #직선
# y=ax+b+error

plt.plot(x, y, 'r') #빨간색 에러들
plt.plot(x, 2 * x + 5, 'k') #검은 직선
plt.show()
```

### np.linspace(start, stop, size)
-  시작값부터 종료값까지 균일한 간격으로 값들을 생성하여 배열로 변환
